```python
import numpy as np
```

```python
tat_td = "tat_Tide_pressure.txt"
tat_td = np.loadtxt(tat_td)
np.mean(tat_td)
```

```python
tat_wh = "tat_Wave_height.txt"
tat_wh = np.loadtxt(tat_wh)
np.mean(tat_wh)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

sns.set(style="whitegrid", font_scale=1.2)  
plt.rcParams['font.weight'] = 'bold'

ax = sns.boxplot(data=[tat_td,tat_wh],orient="h",palette='colorblind')
plt.xlabel('Turnaround time (ms)',weight="bold")
ax.set_yticklabels(['Tide pressure', 'Wave height'],rotation=90, va="center")
plt.tight_layout()
plt.savefig("tat.png")
```
